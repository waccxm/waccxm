```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from plotly.offline import iplot,init_notebook_mode
import plotly.express as px

init_notebook_mode(connected=True)

# plt.rcParams['font.sans-serif']= ['SimHei']
plt.rcParams['axes.unicode_minus']=False

dfModelData=pd.read_csv('./数据分析与应用实验报告模板/labs/model_data.csv',header=0,parse_dates=['model_date','click_date'],dtype={'p_id':'str_'},encoding='gbk')

dfModelData.head()
```

```python
# dfBJModel=dfModelData.groupby(['model_id','p_id','model_date']).sum().reset_index().query("p_id=='90'")
dfJLModelC = dfModelData[(dfModelData['model_id'] == 'C') & (dfModelData['p_name'] == '吉林')]
dfJLModelD = dfModelData[(dfModelData['model_id'] == 'D') & (dfModelData['p_name'] == '吉林')]
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(dfJLModelC['click_date'], dfJLModelC['clicks'], color='blue', label='Model C')
plt.scatter(dfJLModelD['click_date'], dfJLModelD['clicks'], color='red', label='Model D')
plt.title('吉林省C、D模型按点击日期的点击数散点图')
plt.xlabel('点击日期')
plt.ylabel('点击数')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
dfModelData['model_date'] = pd.to_datetime(dfModelData['model_date'])

# 获取所有模型的列表
all_models = dfModelData['model_id'].unique()

# 绘制每个模型的散点图
plt.figure(figsize=(12, 8))
for model in all_models:
    model_data = dfModelData[dfModelData['model_id'] == model]
    plt.scatter(model_data['model_date'], model_data['succs'], label=model)

plt.title('Successes over Model Dates for All Models')
plt.xlabel('Model Date')
plt.ylabel('Successes')
plt.legend()
plt.grid(True)
plt.show()
```

```python
dfModelData['click_date'] = pd.to_datetime(dfModelData['click_date'])

# 筛选出吉林省中模型C和模型D的数据
# 筛选出吉林省中模型C和模型D的数据
dfJLModelC = dfModelData[(dfModelData['model_id'] == 'C') & (dfModelData['p_name'] == '吉林')]
dfJLModelD = dfModelData[(dfModelData['model_id'] == 'D') & (dfModelData['p_name'] == '吉林')]


# 按点击日期分组并求和
dfJLModelC_daily = dfJLModelC.groupby('click_date')['clicks'].sum().reset_index()
dfJLModelD_daily = dfJLModelD.groupby('click_date')['clicks'].sum().reset_index()

# 绘制折线图
plt.figure(figsize=(10, 6))
plt.plot(dfJLModelC_daily['click_date'], dfJLModelC_daily['clicks'], marker='o', color='blue', label='Model C')
plt.plot(dfJLModelD_daily['click_date'], dfJLModelD_daily['clicks'], marker='D', color='red', label='Model D')
plt.title('吉林省C、D模型按点击日期的点击数折线图')
plt.xlabel('点击日期')
plt.ylabel('点击数')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
df_grouped = dfModelData.groupby(['model_id', 'click_date'])['clicks'].sum().reset_index()

# 获取不同模型的数据
model_ids = df_grouped['model_id'].unique()

# 绘制折线图
plt.figure(figsize=(10, 6))
for model_id in model_ids:
    df_model = df_grouped[df_grouped['model_id'] == model_id]
    plt.plot(df_model['click_date'], df_model['clicks'], label=f'Model {model_id}', marker='o')

plt.title('全国每个模型按模型日期的成交数折线图')
plt.xlabel('日期')
plt.ylabel('成交数')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
df_model_A = dfModelData[dfModelData['model_id'] == 'A']

# 对数据进行分组，按照省市进行分组，并对成交数进行求和
df_grouped = df_model_A.groupby('p_name')['clicks'].sum().reset_index()

# 对省市按成交数进行降序排序
df_grouped = df_grouped.sort_values(by='clicks', ascending=False)

# 绘制竖直方向的条形图
plt.figure(figsize=(10, 12))
plt.bar(df_grouped['p_name'], df_grouped['clicks'], color='skyblue')
plt.xlabel('省市')
plt.ylabel('成交数')
plt.title('A模型各省市成交数对比条形图')
plt.xticks(rotation=45)  # 旋转x轴标签，使其更易读
plt.show()
```

```python
fig = px.histogram(data_frame=dfModelData[dfModelData['model_id'] == 'A'], x='clicks', nbins=30,
                   marginal='box', title='A模型成交数直方图', width=1000, height=800)
fig.show()
```

```python
dfModelData['model_id'] = dfModelData['model_id'].apply(lambda s: str(s))
dfModelData['clicks'] = dfModelData['clicks'].apply(lambda s: int(s))
dfModelData['p_name'] = dfModelData['p_name'].apply(lambda s: str(s))

# 筛选出模型为 A 的数据
dfBJModel = dfModelData[dfModelData['model_id'] == 'A']

# 使用 Seaborn 绘制箱线图
plt.figure(figsize=(12, 8))
sns.boxplot(data=dfBJModel, x='p_name', y='clicks')
plt.title('A模型各省市点击数箱线图')
plt.xlabel('省市')
plt.ylabel('点击数')
plt.xticks(rotation=45)
plt.show()
```
